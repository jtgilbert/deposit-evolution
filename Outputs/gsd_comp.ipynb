```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
gsd_ct = {
        "1": 0.005,
        "0": 0.005,
        "-1": 0.0071,
        "-2": 0.0094,
        "-2.5": 0.017,
        "-3": 0.0339,
        "-3.5": 0.051,
        "-4": 0.0675,
        "-4.5": 0.099,
        "-5": 0.128,
        "-5.5": 0.147,
        "-6": 0.1456,
        "-6.5": 0.1214,
        "-7": 0.0983,
        "-7.5": 0.0503,
        "-8": 0.0234
            }
```

```python
count = {float(phi): val / min(gsd_ct.values()) for phi, val in gsd_ct.items()}
```

```python
vol = {phi: ((4/3)*pi*((2**-phi/2)/1000)**3) * val for phi, val in count.items()}
```

```python
vol_gsd = {phi: val / sum(vol.values()) for phi, val in vol.items()}
```

```python
reach_gsd = [val for val in vol_gsd.values()]
reach_gsd.append(0)
dep_gsd = [0.037, 0.037, 0.038, 0.02, 0.02, 0.031, 0.031, 0.038, 0.038, 0.079, 0.079, 0.038, 0.038, 0.2, 0.2, 0.035, 0.035]
vals = [0.5, 1, 2, 4, 5.6, 8, 12, 16, 24, 32, 64, 90, 128, 192, 256, 384, 512]
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(vals, reach_gsd, linewidth=3, c='darkblue', label='Reach')
ax.plot(vals, dep_gsd, linewidth=3, c='darkorange', label='Deposit')
ax.set_xlabel('D (mm)', fontsize=18)
ax.set_ylabel('Fraction', fontsize=18)
ax.tick_params(axis='both', which='major', labelsize=16)
ax.set_xscale('log')
plt.title('Woods GSDs (by volume)', fontsize=22)
plt.legend(fontsize=16)
plt.tight_layout()
plt.savefig('/media/jordan/Elements/Geoscience/Bitterroot/figs/deposit_evolution/woods_gsd_vol.png', dpi=150)
plt.show()
```
